### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
titanic=pd.read_csv('C:/github_repository/2020-python-learning-master/Week4-Pandas/lmk/pandas_exercises_with_data/07_Visualization/Titanic_Desaster/train.csv',encoding='utf-8')
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic=titanic.set_index('PassengerId')
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
male=(titanic['Sex']=='male').sum()
female=(titanic['Sex']=='female').sum()
data=[male,female]
plt.pie(data,labels=['male','female'],autopct='%1.1f%%')
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.stripplot(x='Fare',y='Age',hue='Sex',data=titanic,jitter=True)
```

### Step 7. How many people survived?

### Step 8. Create a histogram with the Fare payed

```python
import numpy as np
fare=titanic.Fare
plt.xlabel('fare')
plt.ylabel('fequency')
plt.hist(fare,bins=np.arange(0,600,10))
```
