### Step 3. Assign it to a variable called chipo.

```python
chipo=pd.read_csv("C:/github_repository/2020-python-learning-master/Week4-Pandas/lmk/pandas_exercises_with_data/data/chipotle.tsv",sep='\t')
```

### Step 4. See the first 10 entries

### Step 5. Create a histogram of the top 5 items bought

```python
top5=chipo.groupby('item_name').quantity.sum().sort_values(ascending=False).iloc[:5]
plt.xlabel('item_name')
plt.ylabel('quantity')
top5.plot(kind='bar')
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
func=lambda x:float(x[1:-1])
chipo.item_price=chipo.item_price.apply(func)
order=chipo.groupby('order_id')[['quantity','item_price']].sum()
plt.xlabel('item_price')
plt.ylabel('order_quantity')
plt.scatter(x=order.item_price,y=order.quantity)
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
chipo['order_price']=chipo['quantity']*chipo['item_price']
order=chipo.groupby('order_id')[['quantity','order_price']].sum()
plt.xlabel('order_price')
plt.ylabel('order_quantity')
plt.scatter(x=order.order_price,y=order.quantity)
```
