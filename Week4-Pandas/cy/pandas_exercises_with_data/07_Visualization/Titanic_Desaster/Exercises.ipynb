### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv('./train.csv')
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId').head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
males = titanic.Sex.value_counts()[0]
females = titanic.Sex.value_counts()[1]
mf_propotion = [males,females]
plt.pie(mf_propotion,labels=['Male','Female'],shadow=False,colors=['yellow','green'],startangle=90,explode=(0,0.15),autopct='%1.2f%%')
plt.axis('equal')
plt.title('the male/female proportion')
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
g = sns.lmplot(x='Age',y='Fare',data=titanic,hue='Sex',fit_reg=False)#regression
g.set(title='Fare payed and the Age')
```

### Step 7. How many people survived?

### Step 8. Create a histogram with the Fare payed

```python
#way 1
#
tF = sns.distplot(titanic.Fare,color='green')
# 
tF.set(xlabel='Fare',ylabel='Frequency',title='Histogram with the Fare payed')
```

```python
# way 2 
plt.rcParams['font.sans-serif']=['KaiTi']
plt.rcParams['axes.unicode_minus']=False
print(titanic.Fare.max())
print(titanic.Fare.min())
b = np.arange(0,520,10)
plt.hist(titanic.Fare,bins=b)
plt.ylim(0,350)
plt.xlim(0,520)
plt.xlabel=('Fare')
plt.ylabel=('Frequency')
plt.title=('Histogram with the Fare payed')
plt.show()
```
