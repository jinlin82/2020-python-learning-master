We define a binary categorial response to predict.

```python
glass['household']=0
for i in range(glass.shape[0]):
    if glass.iloc[i,9]==5|glass.iloc[i,9]==6|glass.iloc[i,9]==7:
        glass.iloc[i,11]=glass.iloc[i,11]+1
print(pd.DataFrame(glass))##DataFrame???
```

Let's change our task, so that we're predicting **household** using **al**. Let's visualize the relationship to figure out how to do this:

```python
plt.scatter(glass.al, glass.household,s=10)
plt.xlabel('al')
plt.ylabel('household')
```

```python
plt.scatter(glass.al, glass.household)
plt.xlabel('al')
plt.ylabel('household')
```

Let's draw a regression line, like we did before:

```python
# fit a linear regression model and store the predictions
x=np.array(glass.al).reshape(-1, 1)
y=np.array(glass.household).reshape(-1, 1)
reg1=LinearRegression().fit(x,y)
glass['hh_pred']=reg1.predict(x)
plt.scatter(x,y,c='b',s=8)
plt.plot(x,glass['hh_pred'],c='r')
plt.xlabel('al')
plt.ylabel('household')
```

```python
# scatter plot that includes the regression line
```

If **al=3**, what class do we predict for household? **1**

If **al=1.5**, what class do we predict for household? **0**

We predict the 0 class for **lower** values of al, and the 1 class for **higher** values of al. What's our cutoff value? Around **al=2**, because that's where the linear regression line crosses the midpoint between predicting class 0 and class 1.

Therefore, we'll say that if **household_pred >= 0.5**, we predict a class of **1**, else we predict a class of **0**.

```python
glass.head()
```

```python
##法一
glass['hh_pred_class']=0
for i in range(glass.shape[0]):
    if glass.iloc[i,12]>=0.5:
        glass.iloc[i,13]=1
    else:
        glass.iloc[i,13]=0
print(glass)
```

```python
plt.scatter(glass.al,glass.household)
glass.al=glass.sort_values(by='al')['al']
glass.hh_pred_class=glass.sort_values(by='al')['hh_pred_class']
plt.plot(al,hh_pred_class,color='red')
plt.xlabel('al')
plt.ylabel('household')
```

```python
import numpy as np
# transform household_pred to 1 or 0
```

```python
##法二
x=np.array(glass.al).reshape(-1, 1)
y=np.array(glass.household).reshape(-1,1)
reg1=LinearRegression().fit(x,y)
glass['household_pred']=reg1.predict(x)
glass['household_pred_class']=np.where(glass.household_pred>=0.5, 1,0)##np.where
glass.head()
```

```python
fig,ax = plt.subplots()
ax.scatter(glass.al, glass.household,edgecolors='k')
ax.set_xlabel('al')
ax.set_ylabel('household')
ax.set_xlim(0.0,4.0)
ax.set_ylim(-0.2,1.2)
ax.tick_params(top='on',right='on',direction='in')
al = glass.sort_values(by='al')['al']
household_pred_class = glass.sort_values(by='al')['household_pred_class']
ax.plot(al,household_pred_class,c='r')
```

## Part 3: Using Logistic Regression Instead

Logistic regression can do just what we did:

```python
# fit a logistic regression model and store the class predictions
from sklearn.linear_model import LogisticRegression
x=np.array(glass.al).reshape(-1, 1)
y=np.array(glass.household).reshape(-1, 1)
reg2=LogisticRegression().fit(x,y)
glass['household_pred']=reg2.predict(x)
glass['household_pred_class'] = np.where(glass['household_pred']>=0.5,1,0)
plt.scatter(x,y,c='b',s=8)
al = glass.sort_values(by='al')['al']
household_pred_class = glass.sort_values(by='al')['household_pred_class']
plt.plot(al,household_pred_class,c='r')
plt.xlabel('al')
plt.ylabel('household')
```

```python
# plot the class predictions
```

What if we wanted the **predicted probabilities** instead of just the **class predictions**, to understand how confident we are in a given prediction?

```python
# store the predicted probabilites of class 1
glass['hh_pred_prob'] = reg2.predict_proba(x)[:, 1]##predict_proba
hh_pred_prob=glass.sort_values(by='al')['hh_pred_prob']
plt.scatter(x, glass.household)
plt.plot(al, hh_pred_prob, color='r')
plt.xlabel('al');plt.ylabel('household')
plt.xlim(0,4);plt.ylim(-0.2,1.2)
```

```python
# plot the predicted probabilities
```

```python
# examine some example predictions
```

```python
print(reg2.predict_proba([[1]]))
print(reg2.predict_proba([[2]]))
print(reg2.predict_proba([[3]]))
```

The first column indicates the predicted probability of **class 0**, and the second column indicates the predicted probability of **class 1**.

### Interpreting Logistic Regression Coefficients

Remember that the probabilities returned by logistic function for class prediction are as follows:
$$p=\frac{e^{\beta_0+\beta_1x}}{1+e^{\beta_0+\beta_1x}}$$

```python
# examine the coefficient for al
```

```python
'al'[0],reg2.coef_[0][0]
```

**Interpretation**: A 1 unit increase in 'al' is associated with a 4.18 unit increase in the log-odds of 'household' (log-odds of A is $log(\frac{P(A)}{1-P(A)})$).

```python
# examine the intercept
```

```python
reg2.intercept_
```

**Interpretation**: For an 'al' value of 0, the log-odds of 'household' is -7.71.

## Part 4: Logistic regression with Titanic data

**Goal**: to predict the survival of the passengers of titanic based on characteristics of passengers ([description](https://raw.githubusercontent.com/o-m-i-d/ML_exercises_2016/master/titanic.names) , [data](https://raw.githubusercontent.com/o-m-i-d/ML_exercises_2016/master/titanic.data))

### Step 1: Read the data into Pandas

```python
import pandas as pd
titanic=pd.read_csv('../titanic.data',index_col='PassengerId')
titanic.head()
```

### Step 2: Define features vectors and response

```python
# chosen features are passenger class and number of parents/children abroad 
features = pd.concat([titanic['Pclass'],titanic['Parch']],axis=1)
X=np.array(features)
y=titanic['Survived']
X;y
```

### Step 3: Split the data into training and testing sets for cross validation

```python
# from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
```

### Step 4: Fit a logistic regression model and obtain the coefficients

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(C=1e9)
logreg.fit(X_train, y_train)
zipped=zip(features.columns,logreg.coef_[0])
list(zipped)
```

### Step 5: Make predictions on the testing set and calculate its accuracy

```python
# class predictions (not predicted probabilities)
y_pred=logreg.predict(X_test)
```

```python
# calculate classification accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### Step 6: Compare the test accuracy with the null accuracy

```python
max(y_test.mean(), 1 - y_test.mean())
```
